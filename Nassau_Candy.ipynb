```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('/content/Nassau Candy Distributor.csv')
```

```python
data.head()
```

```python
data.info()
```

```python
data['Order Date'] = pd.to_datetime(data['Order Date'],errors='coerce').dt.normalize()
data['Ship Date'] = pd.to_datetime(data['Ship Date'],errors='coerce').dt.normalize()
data.info()
```

```python
data['Region'] = data['Region'].str.strip()
data['Region'].unique()
```

```python
data['Order Date'].isna().sum()
```

```python
data['Ship Date'].isna().sum()
```

```python
data['Shipping_lead_time'] = (data['Ship Date'] - data['Order Date']).dt.days
data['Shipping_lead_time'].describe()
```

```python
data_valid = data[data['Shipping_lead_time'] > 0]
```

```python
type(data_valid)
```

```python
avg_lead_time = data_valid['Shipping_lead_time'].mean()
avg_lead_time
```

```python
data_valid['Region_Route'] = "Factory" + "->" + data_valid['Region']
```

```python
data_valid.head()
```

```python
Route_lead_time = data_valid.groupby('Region_Route')['Shipping_lead_time'].mean().sort_values()
Route_lead_time
```

```python
Route_volume = data.groupby('Region').size()
Route_volume
```

```python
data_valid.info()
```

```python
data_valid['State_Route'] = "Factory"+ "->" + data_valid['State/Province']
data_valid.head()
```

```python
ship_mode_summary = data_valid.groupby('Ship Mode')['Shipping_lead_time'].agg(['mean','median','count']).reset_index()
ship_mode_summary
```

```python
State_Route_summary = (data_valid.groupby('State_Route')['Shipping_lead_time'].agg(
    total_shipments=('count'),
    avg_lead_time=('mean'),
    lead_time_variance=('std')
).reset_index())
State_Route_summary.head()
```

```python
Region_Route_summary = (data_valid.groupby('Region_Route')['Shipping_lead_time'].agg(
    total_shipments=('count'),
    avg_lead_time=('mean'),
    lead_time_variance=('std')
).reset_index())
Region_Route_summary.head()
```

```python
Region_Route_summary = Region_Route_summary.sort_values(by='avg_lead_time',ascending=True)
Region_Route_summary
```

```python
State_Route_summary = State_Route_summary.sort_values(by='avg_lead_time',ascending=True)
State_Route_summary.head()
```

```python
top_10_routes = State_Route_summary.head(10)
top_10_routes
```

```python
bottom_10_routes = State_Route_summary.tail(10)
bottom_10_routes
```

```python
geo_summary = (data_valid.groupby('State/Province')['Shipping_lead_time'].agg(
    shipment_volume=('count'),
    avg_lead_time=('mean'),
    variability=('std')
).reset_index())
geo_summary
```

```python
high_lead_time_states = geo_summary.sort_values(by='avg_lead_time',ascending=False).head(10)
high_lead_time_states
```

```python
geo_summary['bottleneck_flag'] = (
    (geo_summary['avg_lead_time'] > geo_summary['avg_lead_time'].mean()) &
    (geo_summary['shipment_volume'] > geo_summary['shipment_volume'].mean())
)
bottleneck_states = geo_summary[geo_summary['bottleneck_flag']]
bottleneck_states
```

```python
ship_mode_comparison = (data_valid.groupby('Ship Mode')['Shipping_lead_time'].agg(
    total_shipments=('count'),
    avg_lead_time=('mean'),
    variability=('std')
).reset_index())
ship_mode_comparison
```

```python
shipmode_cost_profit = (
    data_valid.groupby('Ship Mode').agg(
        avg_lead_time = ('Shipping_lead_time','mean'),
        avg_profit = ('Gross Profit','mean'),
        avg_cost = ('Cost','mean'),
        total_profit = ('Gross Profit','sum'),
        total_cost = ('Cost','sum'),
)
    .reset_index())
shipmode_cost_profit
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
data_valid['efficiency_score'] = 1- scaler.fit_transform(data_valid[['Shipping_lead_time']])
route_efficiency = (data_valid.groupby('Region_Route')['efficiency_score'].mean().reset_index())
route_efficiency
```

```python
route_efficiency = route_efficiency.sort_values(by='efficiency_score',ascending=False)
route_efficiency
```
